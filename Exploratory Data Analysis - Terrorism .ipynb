```python
# Importing Libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading csv file

my_data = pd.read_csv('C:/Users/uzman/Documents/Spark Foundation Internship Project/globalterrorismdb_0718dist.csv',encoding='latin1')
my_data.head()
```

### Performing EDA

```python
# finding shape (rows & column) in our dataset.Which tell us size of data.

my_data.shape
```

```python
my_data.info()
```

```python
# Now we output all columns name

with pd.option_context('display.max_rows', 1, 'display.max_columns', None): 
    display(my_data)
```

```python
# As shown above most of column values are NUll 
# Which is problem to remove these null values to make our data efficient
# What we do is, we select certain column, that are relevant to our project tasks
# Now we are going to select and outcast some of the coulmns
final_dataset =my_data[['iyear','imonth','iday','country_txt','region_txt','provstate','attacktype1_txt',
                                  'targtype1_txt','target1','weaptype1_txt','nkill','nwound','gname','summary']]

# Rename these selected columns to a proper ones
final_dataset.rename(columns={"iyear":"Years",
                                   "imonth":"Months",
                                   "iday":"Days",
                                   "country_txt":"Countries",
                                   "region_txt":"Regions",
                                   "provstate":"States",
                                   "attacktype1_txt":"Attack_types",
                                   "targtype1_txt":"target_types",
                                   "target1":"targets",
                                   "weaptype1_txt":"Weapon_types",
                                   "nkill":"No_of_people_Killed",
                                   "nwound":"No_of_people_Wounded",
                                   "gname":"Groups"},inplace=True)
final_dataset.sample(3)
```

```python
# Here is how it actually summarize 
# below are number of 'counts', 'minimum', 'maxmimim', 'mean', etc

final_dataset.describe()
```

```python
# sub info of dataset
final_dataset.info()
```

### Peforming Data Cleaning 

```python
# To make our data more efficient
# now we work on removing missing values or null values
# this will improve our final output

final_dataset.isnull().sum()
```

```python
# filling missing values from different column

final_dataset['No_of_people_Killed']=final_dataset['No_of_people_Killed'].fillna(0)
final_dataset['No_of_people_Wounded']=final_dataset['No_of_people_Wounded'].fillna(0)
final_dataset['targets']=final_dataset['targets'].fillna('Not known')
final_dataset['summary']=final_dataset['summary'].fillna('Not known')

# Output after altering
final_dataset.isnull().sum()
```

```python
final_dataset.shape
```

```python
# Finding Duplicating values
final_dataset.duplicated().sum()
```

```python
# Droping Duplicating values
final_dataset=final_dataset.drop_duplicates()
```

```python
# The values shows difference after droping duplicate values

final_dataset.shape
```

```python
# Now for the Unique values for each columns
for col in final_dataset:
    print(col,":",final_dataset[col].unique())
    print("."*150)
```

```python
final_dataset['Months'].value_counts()
```

As shown above 13 values and there are 0 as value include also...
Now we trying to replace it with most a month which is most common.

```python
# Output after altering

final_dataset["Months"].mode()
final_dataset['Months']=final_dataset['Months'].replace(0,final_dataset['Months'].mode()[0])
final_dataset['Months'].value_counts()
```

```python
# Total number of months from year (1970-2017) shown in data
# Attack happens in Iraq
total_month=final_dataset.query("Countries=='Iraq'").groupby('Months').mean()
total_month['Days']
# I will directly plot a visual to make it more understandable

plt.figure(figsize=(15,7))
sns.barplot(x=['Jan','Feb','March','April','May','June','July','Aug','Sep','Oct','Nov','Dec'],y=total_month['Days'],data=total_month)
plt.title(" Terrorist Attack in Iraq : 1970-2017",fontsize=25)
plt.xlabel("Months",fontsize=15)
plt.ylabel("Counts",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

The viusal shows Iraq has Average of almost 15 Atacks every Month

Exploring Regions of Iraq, To identify whcih region is most affected!

```python
# Using 'Region' Column we move further
Regions=final_dataset.groupby('Regions')['Years'].count().sort_values(ascending=False).reset_index(name='Count')
Regions
```

```python
# Visuals

plt.figure(figsize=(15,7))
sns.barplot(x='Regions',y='Count',data=Regions)
plt.title(" Terror in different Regions of Iraq : 1970-2017",fontsize=25)
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Counts",fontsize=15)
plt.xticks(rotation=90)

plt.show()
```

The above plot shows Top and Least most affected regions by Terrorism in Iraq.

```python
# Attack Types in Iraq
Attack_type=final_dataset["Attack_types"].value_counts().sort_values(ascending=False).reset_index(name='Count')
Attack_type
```

```python
# Visuals

plt.figure(figsize=(15,8))
sns.barplot(x=Attack_type['index'],y=Attack_type['Count'])
plt.title("Attack_Types in Iraq",fontsize=15)
plt.xlabel("Attack_types",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

Bombing or Explosion is the most used Attack in Iraq. 
Where as Hijacking is the Least common there.

Now let's see the over all visual of attack happen every year👇

```python
years = final_dataset['Years'].unique()
No_of_years = final_dataset['Years'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,7))
sns.barplot(x = years,y = No_of_years)
plt.xticks(rotation = 100)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=20)
plt.title('Years of attack',fontsize=20)
plt.show()
```

As, you can see above 2014 is at peak for attackers.
From 2012 the peak begin to start and in end goes a littel downward.

Below is visuals of Number of Death happen in each year

```python
plt.figure(figsize=(15,8))
final_dataset.groupby(['Years'])['No_of_people_Killed'].sum().plot(kind='bar')
plt.title('Number of Deaths in year',fontsize=25)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Deaths',fontsize=15)
plt.xticks(rotation=100)
plt.show()
```

Again it's Left Skew. Trend rises as move further but slight decrease in death troll in ending years.

```python
# number of deaths taken place in particular region of Iraq.

plt.figure(figsize=(15,8))
final_dataset.groupby(['Regions'])['No_of_people_Killed'].sum().plot(kind='bar')
plt.title('Number of Deaths in Regions',fontsize=25)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Deaths',fontsize=15)

plt.show()
```

Middle East and North Africa is at Top with highest number of Deaths. 

Next we going for Number of wounded people in each Region.

```python
plt.figure(figsize = (15,7))
final_dataset.groupby(['Years'])['No_of_people_Wounded'].sum().plot(kind='bar')
plt.title('Number of Wounded people in years',fontsize=25)
plt.xlabel('Years',fontsize=25)
plt.ylabel('Number of Wounded',fontsize=25)
plt.xticks(rotation=90)
plt.show()
```

```python
# Going deeper, analysis of states of Iraq

plt.figure(figsize=(20,7))
sns.barplot(x=final_dataset['States'].value_counts().head(5).index,y=final_dataset['States'].value_counts().head(5).values,palette="viridis")
plt.title('Top 5 Attacked States of Iraq',fontsize=25)
plt.xlabel('States',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

As we see our Analysis shows Baghdad is state victim of most terrorsit attacks.

```python
# Top 5 weapons used by the terrorists in order for fear to spread:

plt.figure(figsize=(10,6))
sns.barplot(x=final_dataset["Weapon_types"].value_counts()[:5].index,y=final_dataset["Weapon_types"].value_counts()[:5].values,palette="viridis")
plt.title("Percentage of top 5 Weapons used",fontsize=20)
plt.xlabel("Weapons",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

As we see, Explosives are the top weapons used, followed by firearms.

Now listing Organization involve in terrorism.

```python
plt.figure(figsize=(15,10))
sns.barplot(x=final_dataset['Groups'].value_counts()[1:6].values, y=final_dataset['Groups'].value_counts()[1:6].index)
plt.title('Top 5 Terrorist Organizations',fontsize=20)
plt.xlabel('Number of Attacks',fontsize=15)
plt.ylabel('Groups',fontsize=20)
plt.show()
```

Taliban is the first and behind follows ISIL
